- merge concept

```python
a={
    'userid':[1,2,3],
    'name':['sus','bis','gus']
}
b={
    'userid':[1,1,2,4],
    'msg':['theek','bogut','ladukan','boondan']
}
```

```python
person=pd.DataFrame(a)
messgae=pd.DataFrame(b)
```

```python
person
```

```python
messgae
```

```python
person.merge(messgae)
```

```python
person.merge(messgae,how='left')
```

```python
person.merge(messgae,how='right')
```

```python
person.merge(messgae,how='outer')
```
