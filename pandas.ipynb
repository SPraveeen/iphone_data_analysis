# apply method
- takes/input another function 'f'
- it will apply the func f to all the values

```python
def encode(gen):
    if gen=='Male':
        return 1
    else:
        return 0
```

```python
encode('Male')
```

```python
data['gender'].apply(encode)
```

```python
data['gender']=data['gender'].apply(encode)
```

```python
data.head()
```

```python
data[['revenue','budget']]
```

```python
data[['revenue','budget']].apply(np.sum,axis=0)
```

```python
data[['revenue','budget']].apply(np.sum,axis=1)
```

```python
def profit(row):
    return row['revenue']-row['budget']
```

```python
data['profit']=data.apply(profit,axis=1)
```
